```python
# Install Required Libraries
!pip install faker python-docx

from faker import Faker
from docx import Document
import random
import os
import shutil
from google.colab import files

fake = Faker()

# Create folder
os.makedirs("generated_resumes", exist_ok=True)

# Role-wise skills
roles = {

    "Data Analyst": [
        "Python", "SQL", "Power BI",
        "Excel", "Tableau",
        "Statistics", "Data Visualization"
    ],

    "Data Scientist": [
        "Python", "Machine Learning",
        "Deep Learning", "SQL",
        "Pandas", "NumPy",
        "TensorFlow", "Statistics"
    ],

    "Python Developer": [
        "Python", "Flask",
        "Django", "MySQL",
        "Git", "GitHub",
        "REST API", "Streamlit"
    ],

    "Web Developer": [
        "HTML", "CSS",
        "JavaScript", "React",
        "Node.js", "Git",
        "GitHub", "Bootstrap"
    ]
}

# Extra random skills
extra_skills = [

    "Java",
    "C++",
    "MongoDB",
    "AWS",
    "Docker",
    "Linux",
    "Kubernetes",
    "Firebase",
    "Networking",
    "Cyber Security",
    "Data Structures",
    "Operating Systems"
]

# Generate 30 resumes
for i in range(30):

    doc = Document()

    name = fake.name()
    email = fake.email()
    phone = fake.phone_number()

    role = random.choice(
        list(roles.keys())
    )

    role_skills = roles[role]

    # Candidate may miss 1-3 important skills
    num_role_skills = random.randint(
        max(3, len(role_skills)-3),
        len(role_skills)
    )

    selected_role_skills = random.sample(
        role_skills,
        num_role_skills
    )

    # Add random extra skills
    selected_extra_skills = random.sample(
        extra_skills,
        random.randint(2,4)
    )

    final_skills = (
        selected_role_skills
        +
        selected_extra_skills
    )

    random.shuffle(final_skills)

    education = random.choice([
        "B.Tech Computer Science",
        "B.Tech Information Technology",
        "B.Tech Artificial Intelligence and Data Science",
        "B.Sc Computer Science",
        "BCA"
    ])

    experience = random.randint(0,5)

    projects = [

        "Resume Screening System",
        "Student Management System",
        "E-Commerce Website",
        "Weather Forecast Application",
        "Chatbot using NLP",
        "Disease Prediction System",
        "Movie Recommendation System",
        "Sales Dashboard"
    ]

    certifications = [

        "Python Programming",
        "Machine Learning",
        "Power BI",
        "Data Analytics",
        "Web Development",
        "Cloud Computing",
        "SQL Certification"
    ]

    # Resume Content

    doc.add_heading(name, level=1)

    doc.add_paragraph(
        f"Email: {email}"
    )

    doc.add_paragraph(
        f"Phone: {phone}"
    )

    doc.add_heading(
        "Career Objective",
        level=2
    )

    doc.add_paragraph(
        f"Aspiring {role} seeking opportunities to apply technical and analytical skills."
    )

    doc.add_heading(
        "Target Role",
        level=2
    )

    doc.add_paragraph(role)

    doc.add_heading(
        "Education",
        level=2
    )

    doc.add_paragraph(education)

    doc.add_heading(
        "Experience",
        level=2
    )

    doc.add_paragraph(
        f"{experience} Years"
    )

    doc.add_heading(
        "Skills",
        level=2
    )

    for skill in final_skills:
        doc.add_paragraph(skill)

    doc.add_heading(
        "Projects",
        level=2
    )

    for _ in range(2):

        doc.add_paragraph(
            random.choice(projects)
        )

    doc.add_heading(
        "Certifications",
        level=2
    )

    for _ in range(2):

        doc.add_paragraph(
            random.choice(certifications)
        )

    filename = (
        f"generated_resumes/resume_{i+1}.docx"
    )

    doc.save(filename)

print("30 realistic resumes generated successfully!")

# Create ZIP
shutil.make_archive(
    "generated_resumes",
    "zip",
    "generated_resumes"
)

print("ZIP file created!")

# Download ZIP
files.download(
    "generated_resumes.zip"
)
```
